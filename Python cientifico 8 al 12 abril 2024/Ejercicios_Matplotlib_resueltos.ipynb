### Ejercicio 1

### Ejercicio 2

### Ejercicio 3

Genere la gráfica siguiente a partir de los datos proporcionados en la siguiente celda.

![Ejercicio 3 de Matplotlib](imagenes/ejercicios/matplotlib/ejercicio3.jpg)

```python
import matplotlib.pyplot as plt
import numpy as np

# valores para el eje x
dia = np.arange(3, 8)

# valores para el eje y
precio_apertura = np.array([774.25, 776.030029, 779.309998, 779, 779.659973])
precio_maximo = np.array([776.065002, 778.710022, 782.070007, 780.47998, 779.659973])
precio_minimo = np.array([769.5, 772.890015, 775.650024, 775.539978, 770.75])
precio_cierre = np.array([772.559998, 776.429993, 776.469971, 776.859985, 775.080017])

# Resuelva el ejercicio en las siguientes líneas de esta celda
fig, ax = plt.subplots()

# gráficas de líneas 
ax.plot(dia, precio_apertura, label='Apertura')
ax.plot(dia, precio_maximo, label='Máximo')
ax.plot(dia, precio_minimo, label='Mínimo')
ax.plot(dia, precio_cierre, label='Cierre')

# Nombre para el eje x
ax.set_xlabel('Día')
# Nombre para el eje y
ax.set_ylabel('Cotización')

# Título
ax.set_title('Cotizaciones bancarias diarias')

# Leyenda
ax.legend(loc='lower center')

# Muestro la malla
ax.grid()

# Mostramos el gráfico en el notebook
# plt.show()
fig.savefig('imagenes/ejercicios/matplotlib/ejercicio3.jpg')
```

### Ejercicio 5

A partir de los datos de la siguiente celda, reproduzca el gráfico que se muestra a continuación.

![Ejercicio 5 de Matplotlib](imagenes/ejercicios/matplotlib/ejercicio5.jpg)

```python
from matplotlib import pyplot as plt
import numpy as np

dia = np.arange(3, 8)
precio_bolsa = np.array([772.559998, 776.429993, 776.469971, 776.859985, 775.080017])

# Resuelva el ejercicio en las siguientes líneas de esta celda
fig, ax = plt.subplots()

# Plot the data as a red line with round markers
ax.plot(dia, precio_bolsa, color='red', marker='o')

# Set the xtick locations
ax.set_xticks(dia)

# Set the xtick labels
# ax.set_xticklabels(['Dia '+str d for d in dia])

# Etiqueta para el eje x
plt.xlabel('Fecha de interés')
# Etiqueta para el eje y
plt.ylabel('Precio de cierre')
# Título
plt.title('Precio de cierre de los primero días') 

# Establecemos las posiciones de las marcas visibles auxiliares en posiciones concretas para el eje X
ax.set_yticks(np.arange(772, 777.5, 0.1), minor=True)

# una malla personalizada con origen en las marcas auxiliares para el eje X 
ax.grid(which='minor', axis='y', color='gray', linestyle=':', linewidth=0.5)
# Malla personalizada para los major ticks
plt.grid(which='major', linestyle='-',  color='black', linewidth=0.5)

# Muestro la gráfica
plt.show()
#fig.savefig('imagenes/ejercicios/matplotlib/ejercicio5.jpg')
```

### Ejercicio 6

Genere el gráfico que se muestra a continuación a partir de los datos que se proporcionan en la siguiente celda.

![Ejercicio 6 de Matplotlib](imagenes/ejercicios/matplotlib/ejercicio6.jpg)

```python
import matplotlib.pyplot as plt

# Data para mostrar
lenguajes = 'Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'
popularidad = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

# Escriba su solución en las siguientes líneas de código

# generamos figura y ejes asociados
fig, ax = plt.subplots()

# resaltamos 'Python'
resalte = (0, 0.1, 0, 0,0,0)  
# gráfica
ax.pie(popularidad, explode=resalte, labels=lenguajes, \
       autopct='%1.1f%%', shadow=True)

# Mostramos el resultado
plt.show()
fig.savefig('imagenes/ejercicios/matplotlib/ejercicio6.jpg')
```

### Ejercicio 7

Genere el gráfico de barras que se muestra a partir de la información que se proporciona en la siguiente celda.

![Ejercicio 7 de Matplotlib](imagenes/ejercicios/matplotlib/ejercicio7.jpg)

```python
import matplotlib.pyplot as plt
import numpy as np

etiquetas = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularidad = np.array([22.2, 17.6, 8.8, 8, 7.7, 6.7])

# Escriba su solución en las siguientes líneas de esta celda

# Genero figura y ejes asociados
fig, ax = plt.subplots()

x = np.arange(popularidad.size)
ax.bar(x, popularidad, color='green', alpha=0.5)
ax.set_xlabel("Lenguajes de programación")
ax.set_ylabel("Popularidad (%)")
ax.set_title("popularidad de lenguajes\nInformación extraida de CrazyDevelopers")
ax.set_xticks(x, etiquetas)
# Mostramos el gráfico
plt.show()
# fig.savefig('imagenes/ejercicios/matplotlib/ejercicio7.jpg')
```

### Ejercicio 8

A partir de los datos proporcionados en la siguiente celda, genere un gráfico como el que se muestra a continuación.

![Ejercicio 8 de Matplotlib](imagenes/ejercicios/matplotlib/ejercicio8.jpg)

```python
import matplotlib.pyplot as plt 
import numpy as np 

# Creamos un generador
rng2 = np.random.default_rng(seed = 123456)

# Muestras aleatorias de coordenadas x e y con distribución normal
x = rng2.normal(3, 3, 50)
y = rng2.normal(5, 1, 50)

# Escriba la solución en las siguiente líneas de esta celda


# Generación de figura y ejes asociados
fig, ax = plt.subplots()

# Diagrama de dispersión
ax.scatter(x, y, s=70, facecolors='none', edgecolors='g')

# Etiquetas en los ejes
ax.set_xlabel("X")
ax.set_ylabel("Y")

# Mostrar el gráfico
plt.show()
# fig.savefig('imagenes/ejercicios/matplotlib/ejercicio8.jpg')
```
